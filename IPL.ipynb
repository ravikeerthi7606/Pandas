```python
plt.figure(figsize=(10,5))
matches["winner"].value_counts().head(10).plot(kind="bar")
plt.title("Top Winning Teams")
plt.show()
```

```python
plt.figure(figsize=(10,5))
matches["toss_decision"].value_counts().plot(kind="bar")
plt.title("Toss Decisions")
plt.show()
```

```python
plt.figure(figsize=(10,5))
matches["player_of_match"].value_counts().head(20).plot(kind="bar")
plt.title("Player of the Match Awards")
plt.show()
```

```python
top_runs = deliveries.groupby("batter")["batsman_runs"].sum()
```

```python
deliveries.columns
```

```python
top_runs.sort_values(ascending=False).head(10)
```

```python
top_runs.sort_values(ascending=False).head(10).plot(kind="bar")
plt.title("Top 10 Run Scorers")
plt.show()
```

```python
wickets = deliveries[deliveries['is_wicket']==1]
```

```python
wickets['bowler'].value_counts().head(10)
```

```python
wickets['bowler'].value_counts().head(10).plot(kind='bar')
plt.title("top wicket-taking bowler")
plt.show()
```

```python
matches["toss_decision"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.ylabel("")
plt.show()
```

```python
fours = deliveries[deliveries["batsman_runs"]==4]
fours.groupby("batter")["batsman_runs"].count().sort_values(ascending=False).head(10)
```

```python
six = deliveries[deliveries["batsman_runs"]==6]
six.groupby("batter")["batsman_runs"].count().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x=matches['result'])
plt.title('Match Result Types')
plt.show()
```

```python
team_wins = matches['winner'].value_counts()
total_matches = matches['team1'].value_counts() + matches['team2'].value_counts()
win_percentage = (team_wins / total_matches) * 100
win_percentage.sort_values(ascending=False)
```

```python
numeric_columns = deliveries.select_dtypes(include=np.number)
plt.figure(figsize=(12,8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.histplot(deliveries['total_runs'])
plt.title('Distribution of Runs')
plt.show()
```
